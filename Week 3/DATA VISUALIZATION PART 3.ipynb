```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import warnings
```

```python
warnings.simplefilter("ignore")
```

```python
# Barplot
sns.barplot(x = 'day', y='tip', data=tips, hue='smoker')
sns.despine()
```

```python
# Barplot
sns.barplot(x = 'day', y='tip', data=tips, hue='smoker',loc='top')
sns.despine(left=True)
```

```python
# Barplot
sns.barplot(x = 'tip', y='day', data=tips, hue='smoker',ci=None,palette='summer')
sns.despine()
```

```python
# BOXPLOT
plt.figure(figsize=(12,8))
sns.boxplot(x='day', y='total_bill',hue='smoker', data=tips)
```

```python
# BOXPLOT
plt.figure(figsize=(12,8))
sns.boxplot(x='day', y='tip',hue='smoker', data=tips)
sns.despine()
```

```python
# ViolinPlot
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# ViolinPlot
sns.violinplot(x="day", y="total_bill", data=tips,palette='Set2', hue='sex')
```

```python
# ViolinPlot
sns.violinplot(x="day", y="total_bill", data=tips,palette='Set3', hue='smoker',split=True)
```
