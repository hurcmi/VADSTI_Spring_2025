```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Wikipedia URL to scrape
url = "https://en.wikipedia.org/wiki/Ghana"

# Set user-agent to avoid request blocking
headers = {"User-Agent": "Mozilla/5.0"}

# Send HTTP request
response = requests.get(url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract page title
title = soup.find("h1").text

# Extract key historical events from the first few paragraphs
paragraphs = soup.find_all("p", limit=5)
important_events = "\n".join([p.text.strip() for p in paragraphs])

# Extract key facts from the infobox
infobox = soup.find("table", {"class": "infobox"})
infobox_data = {}

if infobox:
    rows = infobox.find_all("tr")
    for row in rows:
        header = row.find("th")
        data = row.find("td")
        if header and data:
            infobox_data[header.text.strip()] = data.text.strip()

# Convert extracted data into a Pandas DataFrame
data = {
    "Page Title": [title],
    "Important Events": [important_events]
}

# Add infobox data to the DataFrame
for key, value in infobox_data.items():
    data[key] = [value]

df = pd.DataFrame(data)

# Save DataFrame to a CSV file
csv_filename = "ghana_wikipedia_data.csv"
df.to_csv(csv_filename, index=False, encoding="utf-8")

print(f"Data successfully saved to {csv_filename}")

# Display the DataFrame
print(df)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL to scrape
url = "https://user.iiasa.ac.at/~marek/fbook/01/geos/gh.html"

# Send a GET request
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract the title of the page
title = soup.find("title").text if soup.find("title") else "No title found"

# Extract key information from paragraphs
paragraphs = soup.find_all("p")
important_facts = "\n".join([p.text.strip() for p in paragraphs[:5]])  # Extract first 5 paragraphs

# Extract tabular data if available
tables = soup.find_all("table")
table_data = {}

if tables:
    for table in tables:
        rows = table.find_all("tr")
        for row in rows:
            cells = row.find_all(["th", "td"])
            if len(cells) == 2:
                key = cells[0].text.strip()
                value = cells[1].text.strip()
                table_data[key] = value

# Convert extracted data into a Pandas DataFrame
data = {
    "Page Title": [title],
    "Important Facts": [important_facts]
}

# Add tabular data to the DataFrame
for key, value in table_data.items():
    data[key] = [value]

df = pd.DataFrame(data)

# Save DataFrame to a CSV file
csv_filename = "ghana_facts.csv"
df.to_csv(csv_filename, index=False, encoding="utf-8")

print(f"Data successfully saved to {csv_filename}")

# Display the DataFrame
df.head(20)
```
