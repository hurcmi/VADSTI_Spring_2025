# DATA VISUALIZATION PART 2


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import warnings
```

```python
warnings.simplefilter('ignore')
```

## Lineplot

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.show()
```

## Countplot

```python
sns.countplot(x='day',hue='sex',data=tips)
```

```python
sns.countplot(x='day',hue='smoker',data=tips)
```

## Barplot

```python
plt.figure(figsize=(12,8))
sns.barplot(data=tips, x='tip',y='total_bill', hue='sex')
plt.tight_layout()
```

## Histogramplot

```python
sns.histplot(tips['total_bill'], color='orange',edgecolor='blue',lw=3)
```

```python
sns.histplot(tips['total_bill'], color='orange',edgecolor='purple',lw=3,ls='--')
sns.despine()
```

```python
sns.histplot(tips['total_bill'], color='aliceblue',edgecolor='k',lw=2)
sns.despine(left=True)
```

## JointPlot

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex',color='crimson')
```

## Pairplot

```python
sns.pairplot(tips)
```
