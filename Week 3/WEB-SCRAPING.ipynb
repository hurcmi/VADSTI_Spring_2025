```python
import requests
import re
import spacy
from bs4 import BeautifulSoup

# Load spaCy NLP model
nlp = spacy.load("en_core_web_sm")

# Wikipedia API URL
wiki_api_url = "https://en.wikipedia.org/w/api.php"

# Parameters for the API request
params = {
    "action": "query",
    "titles": "United States Military Academy",
    "prop": "revisions",
    "rvprop": "content",
    "format": "json"
}

# Make API request
response = requests.get(wiki_api_url, params=params)
data = response.json()

# Extract page content
pages = data.get("query", {}).get("pages", {})
page_id = list(pages.keys())[0]

if "revisions" in pages[page_id]:
    page_content = pages[page_id]["revisions"][0]["*"]

    # Remove unwanted characters
    cleaned_content = re.sub(r"[{}/\\]", "", page_content)

    # Use BeautifulSoup with html5lib to clean the content further
    soup = BeautifulSoup(cleaned_content, "html.parser")
    text_content = soup.get_text()

    # Process text with spaCy for Named Entity Recognition (NER)
    doc = nlp(text_content)

    # Extract important date-related information
    extracted_dates = []
    for ent in doc.ents:
        if ent.label_ == "DATE":
            extracted_dates.append(ent.text)

    # Get unique dates
    extracted_dates = list(set(extracted_dates))
    extracted_dates.sort()

    # Extracting relevant historical facts
    historical_events = {
        "First Women Cohort Graduated": None,
        "First African American Admitted": None,
        "Academy Established": None
    }

    # Searching for relevant historical facts
    for sentence in doc.sents:
        sentence_text = sentence.text.lower()
        
        if "first women" in sentence_text and "graduate" in sentence_text:
            historical_events["First Women Cohort Graduated"] = sentence.text

        if "first african american" in sentence_text and "admitted" in sentence_text:
            historical_events["First African American Admitted"] = sentence.text

        if "established" in sentence_text and "1802" in sentence_text:
            historical_events["Academy Established"] = sentence.text

    # Display extracted data
    print("===== Important Dates Extracted from Wikipedia =====")
    for date in extracted_dates:
        print(f"- {date}")

    print("\n===== Key Historical Events at West Point =====")
    for key, value in historical_events.items():
        if value:
            print(f"{key}: {value}")
        else:
            print(f"{key}: Information not found in extracted text.")

else:
    print("No content found for the specified page.")
```

```python
text = "Mike likes baseball. Mike enjoys running"
```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"POS" : "PROPN", "OP" : "+"}]
matcher.add("PROPER_NOUN", [pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
matches.sort(key=lambda x:[1])
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
import requests
import json

# Define the URL for the MediaWiki API
wiki_api_url = "https://en.wikipedia.org/w/api.php"

# Specify parameters for the API request
params = {
    "action": "parse",
    "page": "Battle_of_Midway",
    "prop": "wikitext",
    "format": "json"
}

# Make the API request
response = requests.get(wiki_api_url, params=params)
data = response.json()

# Extract the page content from the JSON response
page_content = data["parse"]["wikitext"]["*"]

# Save the page content to a JSON file
with open("Battle_of_Midway.json", "w") as json_file:
    json.dump({"page_content": page_content}, json_file, indent=4)

print("Page content saved to 'Battle_of_Midway.json'.")
```

```python
import requests
from bs4 import BeautifulSoup
import json

# Define the URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Battle_of_Midway"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")

# Find the main content of the page
content_div = soup.find(id="mw-content-text")

# Extract text from paragraphs within the main content
paragraphs = content_div.find_all("p")

# Concatenate the text from all paragraphs
page_text = "\n".join([p.get_text() for p in paragraphs])

# Save the extracted text to a JSON file
with open("Battle_of_Midway_text.json", "w") as json_file:
    json.dump({"page_text": page_text}, json_file, indent=4)

print("Text content saved to 'Battle_of_Midway_text.json'.")
```
