# What is Unsupervised Learning?

Unsupervised learning is a type of machine learning where the model learns from the data without any guidance. In other words, there are no explicit labels or target outcomes provided to the model. The goal of unsupervised learning is to identify patterns, relationships, or structure in the input data.

There are two main types of unsupervised learning algorithms:

1. **Clustering Algorithms**: These algorithms group data based on inherent similarities. They are used to explore and understand the structure or pattern within a dataset. Examples include k-Means clustering, Hierarchical clustering, and DBSCAN.

2. **Dimensionality Reduction Algorithms**: These algorithms are used to reduce the number of variables in a dataset while preserving the essential features. They are useful for visualizing high-dimensional data and improving the efficiency of other machine learning models. Examples include Principal Component Analysis (PCA) and t-SNE.

Unsupervised learning can be quite powerful when used correctly, as it can reveal hidden patterns and correlations in the data that might not be apparent or intuitive. However, the results can also be more challenging to interpret and validate because there's no "ground truth" to compare them against.

### Unsupervised Learning Analogy: Organizing books

Let's consider a real-world example: the task of organizing a large library full of different types of books that are all mixed together.

If you were given a list of categories and asked to organize the books based on those categories (for instance, fiction, non-fiction, science, literature, etc.), that would be like supervised learning. You have labels (the categories) and you're asked to classify the books according to these labels.

However, if no categories were provided and you were simply asked to arrange the books in a way that made sense, you would have to figure out your own system for organization. You might start noticing patterns or similarities among the books - maybe you would group them by size, by color of their covers, or perhaps you'd read a few pages and try to group them by topic or writing style. This process of finding inherent patterns and structures in the data without having any predefined labels is similar to unsupervised learning.

In this case, your task is similar to a clustering algorithm in unsupervised learning, where the aim is to find inherent groupings within the data. The groups or 'clusters' you create would be based on the patterns and relationships you've identified in the data, even though no one explicitly told you what to look for.

### What is Clustering?

* Clustering models are a type of unsupervised learning technique used to identify patterns or groupings in data without the use of predefined labels.

* The goal of clustering is to partition the dataset into groups (or "clusters") in such a way that data points in the same cluster are more similar to each other than to those in other clusters.

* Clustering is often used in exploratory data analysis to understand the underlying structure of the data, discover patterns and detect outliers.

#### Advantages of Clustering

Clustering, being a key unsupervised learning technique, has several advantages and disadvantages:

### Advantages of Clustering:

1. **Unsupervised Classification**: Clustering can be seen as an unsupervised classification method. It allows users to perform classification without having prior labels for the groups.

2. **Data Summarization**: Clustering algorithms can help in data reduction by representing large datasets by a smaller number of representative clusters.

3. **Pattern Discovery**: Clustering can help discover hidden patterns in the data that might not be obvious or detectable by other techniques.

4. **Feature Engineering**: New features can be created based on clustering, which can be used to boost the performance of supervised learning algorithms.

5. **Market Segmentation**: In business applications, clustering helps in market segmentation; categorizing customers based on purchasing behavior, preferences, or other metrics.

6. **Image Segmentation and Anomaly Detection**: Clustering can be applied in image processing for image segmentation and in security for anomaly detection.

### Disadvantages of Clustering:

1. **Sensitivity to Initial Conditions**: Algorithms like k-means can produce different clustering results if the initial conditions change.

2. **Difficulty in Choosing Appropriate Clusters**: It's often hard to determine the number of clusters (like k in k-means). Methods such as the elbow method or silhouette score can help but might not always provide a clear answer.

3. **Assumption of Cluster Shapes**: Some clustering algorithms assume clusters to be spherical, which may not always be the case. For instance, k-means tends to find spherical clusters.

4. **Scaling of Data**: Many clustering algorithms are sensitive to the scale of the data. Data often needs to be normalized or standardized for better results.

5. **High Dimensionality**: Clustering in high-dimensional spaces can be problematic. The distance between points tends to become uniform, making clustering challenging. Techniques like dimensionality reduction may be needed.

6. **Noisy Data**: The presence of noise and outliers in data can lead to the formation of artificial clusters which might not have any real significance.

Understanding these pros and cons is essential when considering the use of clustering algorithms for specific tasks.

### Steps for defining Clustering Model Tasks

The unsupervised learning process involves several steps, similar to other types of machine learning. Here is a breakdown:

1. **Data Collection**: This involves gathering data relevant to the problem you are trying to solve. The data in unsupervised learning doesn't have target labels or outcomes.

2. **Data Preprocessing**: Once you have the data, you need to clean and preprocess it. This can involve dealing with missing values, normalizing numerical data (so all features have a similar scale), and encoding categorical data.

3. **Feature Selection/Extraction**: While not always necessary, this step can be useful when dealing with high-dimensional data. Techniques like Principal Component Analysis (PCA) can reduce the dimensionality of the data, making it easier to work with.

4. **Model Selection**: Choose the type of unsupervised learning algorithm you want to use. The choice depends on the problem at hand and the nature of your data.

5. **Training the Model**: Feed your data into the model. The model will attempt to find patterns or structures in the data, like clusters in a clustering algorithm.

6. **Interpreting the Results**: In unsupervised learning, this can be more challenging since there are no true labels to compare the model's output against. The interpretation often depends on domain knowledge and the specific problem you are trying to solve.

7. **Evaluation**: In unsupervised learning, evaluation is often subjective and depends on whether the patterns found by the model are useful or meaningful in the specific context.

8. **Model Tuning/Improvement**: Based on the results and any performance metrics you might have, you can go back and tweak the model or preprocessing steps to try and improve the results.

Remember, unsupervised learning is typically used for exploratory purposes to find hidden patterns or structures in data. The results need careful interpretation and often serve as a starting point for further analysis.

Clustering models can be broadly classified into several types based on their approach to creating clusters:

1. **Centroid-based Clustering**: These algorithms represent each cluster by a single mean vector. The mean vector could be the mean of feature vectors of all the objects in the cluster. The most common centroid-based clustering technique is **K-means**.

2. **Connectivity-based Clustering (Hierarchical Clustering)**: These algorithms build clusters based on the principle of connectivity. The basic idea is that objects that are closer in the feature space are more related (or similar) than those farther away. Hierarchical clustering builds a multilevel hierarchy of clusters by creating a cluster tree.

3. **Distribution-based Clustering**: These algorithms assume that the data is composed of distributions, such as Gaussian distributions. As a result, objects in the same cluster are part of the same distribution. A popular example of distribution-based clustering is **Gaussian Mixture Models (GMM)**.

4. **Density-based Clustering**: These algorithms find the areas of high density and separate them from areas of low density. They are especially useful when dealing with spatial data or when there is noise in your data. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** is a widely-used density-based clustering algorithm.

## K-means Clustering

K-means is a centroid-based or partitioning method of clustering. It's one of the simplest unsupervised learning algorithms that solve the well-known clustering problem.  The aim is to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean (cluster centers or cluster centroid), serving as a prototype of the cluster.

#### Algorithm Description:
- **Step 1:** K initial "means" (centroids) are randomly generated within the data domain.
- **Step 2:** K clusters are created by associating every observation with the nearest centroid.
    - The distance between the datapoints and the centroid is calculated using a distance metric usch as Euclidian disance
- **Step 3:** The centroid of the clusters becomes the new mean.
- **Step 4:** Steps 2 and 3 are repeated until the centroids do not change significantly, or a maximum number of iterations is reached.

#### Advantages:
- Fast and efficient in terms of computational cost.
- Easy to implement.

#### Disadvantages:
- Number of clusters, K, needs to be defined beforehand.
- The initial choice of centroids can affect the output (this problem is often remedied with methods like K-means++).
- K-means assumes that clusters are convex and isotropic, which is not always the case. It can be impacted by the scale of the dataset.
- Not suitable for clusters of different sizes and densities.

#### Hands-On Example: Implementing K-means using Sckit-learn

Let's use the `breast cancer` dataset available in `sklearn.datasets` to demonstrate a K-means clustering task.

```python
# Import necessary libraries
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the breast cancer dataset
data = datasets.load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
# It's a good practice to scale our data for better results
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled
```

```python
# Define the K-means clustering model
kmeans = KMeans(n_clusters=2, random_state=42)  # We choose 2 clusters as there are 2 classes in the breast cancer dataset.

# Fit the model to the data
kmeans.fit(df_scaled)

# Get the cluster labels for each sample
labels = kmeans.labels_

# Plot the first two features with colors representing their labels
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Visualization of KMeans Clustering for Breast Cancer data')
plt.show()
```

In this example, we load the breast cancer dataset, define a K-means model with 2 clusters, fit the model to the data, and retrieve the cluster labels. The resulting plot shows the data points colored according to their cluster label, with the red dots representing the cluster centers.

Remember, the real breast cancer data has labels (it's a supervised learning problem). However, in practice, we would use clustering when we do not have labels. We are only using this dataset for simplicity and visualization.

### Hierarchical Clustering

**Hierarchical Clustering** is another unsupervised learning algorithm that is used to group together the unlabeled datasets into a cluster, and also described as a hierarchy, similar to a tree structure.

The algorithm starts by treating each object as a singleton cluster. Next, pairs of clusters are successively merged until all clusters have been merged into a single cluster that contains all objects. The result is a tree-based representation of the objects, named **dendrogram**.

More generally, the hierarchical clustering technique calculates the distance between each of the data points and assigns it to the nearest cluster. It's a bottom-up approach: each data point is treated as a single cluster initially, and then successively merged into bigger clusters.

**What is a dendrogram?**
A dendrogram is a tree-like diagram that displays the hierarchical relationship between objects. It is most commonly used in hierarchical clustering and is a visualization tool for the clustering processes. Dendrograms are used for visualizing the history of groupings and understanding the cluster level distances.

<!--  -->


Here's how to interpret a dendrogram:

1. Each leaf of the dendrogram represents one data point. At the bottom, we have as many clusters as data points.

2. As you move up the y-axis of the dendrogram, clusters merge and we get fewer and larger clusters. The height at which two branches merge represents the distance between the two clusters being merged.

3. At the top of the dendrogram, all data points merge into a single cluster.

4. If you draw a horizontal line anywhere on the dendrogram, the number of vertical lines it crosses gives you the number of clusters at that height.

5. The distance between clusters can be Euclidean distance, Manhattan distance, Cosine similarity, etc., depending on the method used for clustering.

In hierarchical clustering, the dendrogram is used to decide the number of clusters by setting a threshold distance and cutting the dendrogram at that level. The number of disconnected branches at a given height are the number of clusters at that distance threshold.


**Pros:**

- **Flexibility with Clusters:** No need to specify number of clusters upfront; it can be decided by looking at the dendrogram.
- **Insightful Visual Representation:** Dendrogram offers a detailed view of clusters and their relationships.
- **Detailed Information:** Provides hierarchical organization of the data beyond just the final clusters.

**Cons:**

- **High Computational Cost:** Time complexity ranges from O(n^2) to O(n^3), making it inefficient for large datasets.
- **Irreversible Decisions:** Once two clusters are merged, it can't be undone; no reallocation of instances is allowed.
- **Sensitive to Noise and Outliers:** Outliers can affect the merging process, leading to sub-optimal clusters.
- **Difficulty with Variably Sized/Dense Clusters:** Struggles with clusters of different sizes, densities, and non-globular shapes.


**Difference between Hierarchical Clustering and K-means**
It does not require us to pre-specify the number of clusters to be generated as is required by the k-means approach. Furthermore, hierarchical clustering provides a tree-based representation of the observations, called a dendrogram.

#### Hands-On Example: Implementing Hierarchical using Sckit-learn

Here's an example of how we can implement Hierarchical Clustering:

```python
# Import the necessary libraries
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

# Load the breast cancer dataset
data = datasets.load_breast_cancer()
X = data.data[:30]
y = data.target[:30]

# It's a good practice to scale our data for better results
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Use the 'ward' method for linkage
linked = linkage(X_scaled, 'ward')

# Plot the hierarchical clustering as a dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()
```

The numbers on a dendrogram represent the distance or dissimilarity between clusters.

1. **Y-axis**: The y-axis of a dendrogram represents the level of dissimilarity or distance between clusters. The height of the branches represents the distance between the clusters being merged. The higher the branch, the greater the dissimilarity between the clusters, which means the clusters are more different from each other.

2. **X-axis**: The x-axis of a dendrogram represents the samples or clusters. Each leaf on the dendrogram represents one of the data points, and the horizontal lines joining them show the points being grouped together into a cluster.

The distances are determined by the method of distance measurement and linkage specified in the clustering process. The linkage criteria determine the distance between newly formed clusters, and different linkage methods have different ways of calculating this.

In this code snippet above, we use the 'ward' method as the linkage criteria. The Ward method minimizes the sum of squared differences within all clusters (variance-based approach). It is a minimization algorithm to minimize the total within-cluster variance. Other common linkage methods include 'complete', 'average', and 'single'.

Remember, in practice, hierarchical clustering is not usually suitable for large datasets.

### Density-Based Spatial Clustering of Applications with Noise (DB-SCAN)

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a popular clustering algorithm that differentiates itself from methods like k-means by not assuming that clusters are necessarily spherical and by being able to identify clusters of arbitrary shapes. It also has the capability to detect noise in the data.


**Basics of DBSCAN:**
1. **Density**: The main idea behind DBSCAN is that a cluster is a dense region of points, which is separated from other dense regions by areas with low density of points.

2. **Core Points, Border Points, and Noise**:
   - **Core Point**: A point is a core point if there are at least `min_samples` points within a distance of `eps` from it (including the point itself).
   - **Border Point**: A point that isn't a core point but is within the `eps` distance from a core point.
   - **Noise**: Points that are neither core points nor border points.

3. **Algorithm Process**:
   - Randomly pick a point not already visited.
   - If there are `min_samples` points within `eps` distance of this point, a new cluster is formed, and the cluster is grown by adding all points within `eps` distance of the original point (and all their `eps` neighbors, and so on).
   - If the point is within `eps` distance of only fewer than `min_samples` points, it's labeled as noise (this could change later if the point is found within `eps` distance from a different core point).
   - Repeat the process until all points are either assigned to a cluster or labeled as noise.

**Advantages:**
- Can find arbitrarily shaped clusters.
- Doesn’t require the number of clusters to be specified in advance.
- Able to handle noise.
- Doesn’t require clusters to be of similar size.

**Disadvantages:**
- Doesn’t perform well with clusters of varying densities.
- The quality of the clustering largely depends on the correct choice of `eps` and `min_samples`.
- Not well-suited for high-dimensional data.
  
**Parameters to tune**:
1. **`eps` (epsilon)**: It's the distance within which points are considered neighbors. A small `eps` value means that clusters will be closer together, while a large `eps` can result in merging of clusters that should be separate.
  
2. **`min_samples`**: The minimum number of points required to form a dense region. For larger datasets, it may be necessary to increase this number.

**Applications**:
DBSCAN is widely used in anomaly detection, where data can be largely unstructured, and we are more interested in the dense regions and would like to treat sparse regions as noise or anomalies.

Overall, DBSCAN is an excellent tool when you expect your data to have clusters of similar density, and those clusters might not necessarily be globular in shape. However, like all models, understanding the underlying assumptions and the nature of your data is key to using it effectively.

#### Hands-On Example: Implementing DBSCAN using Sckit-learn

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Load the dataset
data = datasets.load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Scale the data for better results with DBSCAN
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# DBSCAN Clustering
# These parameters (eps and min_samples) are just starting points.
# You'll often need to adjust them based on the nature of your dataset.
dbscan = DBSCAN(eps=2, min_samples=10)
clusters = dbscan.fit_predict(df_scaled)

# Visualize the clustering result
# For visualization, let's use the first two principal components
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_scaled)
principalDf = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])

plt.figure(figsize=(8,6))
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'], c=clusters, cmap='rainbow')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('DBSCAN Clustering using 2 Principal Components')
plt.show()
```

### Evaluationg Clustering Models

Evaluating clustering models can be a challenging task due to the unsupervised nature of clustering. However, various metrics and techniques have been developed to help in the assessment of clustering results. The choice of metric often depends on the specific problem, the data, and the goals of the clustering task. Below are some common methods and metrics used for clustering evaluation:

1. **Internal Evaluation Metrics** (don't require ground truth labels):
   
   - **Silhouette Coefficient**: Measures how close each point in one cluster is to the points in the neighboring clusters. Its values range from -1 to 1, where a high value indicates good clustering.
   
   - **Davies-Bouldin Index**: It measures the average similarity ratio of each cluster with its most similar cluster. Lower values indicate better clustering.
   
   - **Within-Cluster-Sum-of-Squares (WCSS)**: Specifically used for K-means. It measures the squared average distance of all points within a cluster to the centroid of the cluster. The goal is to minimize this value.

   - **Dunn Index**: It compares the sizes and distances of clusters. It's the ratio of the smallest distance between points in different clusters to the largest intra-cluster distance.

2. **External Evaluation Metrics** (require ground truth labels):

   - **Adjusted Rand Index (ARI)**: Measures the similarity of two assignments (true labels vs. predicted clusters) after correcting for chance. It ranges from -1 to 1, where a high value indicates that clustering is consistent with the ground truth.
   
   - **Normalized Mutual Information (NMI)**: Measures the mutual information shared between the true labels and the cluster assignments, normalized to have values between 0 and 1.
   
   - **Fowlkes-Mallows Index (FMI)**: It's the geometric mean of precision and recall.

   - **Jaccard Index**: Measures the proportion of shared members between true and predicted clusters in relation to the size of the union of the two sets.

3. **Visual Evaluation**:

   - **Scatter Plots**: Especially when data is 2D or can be reduced to 2D using techniques like PCA or t-SNE, visualizing clusters can be effective.

   - **Dendrogram**: For hierarchical clustering, dendrograms can be used to visualize the formation of clusters at various levels of similarity.

5. **Comparing Different Clustering Algorithms**:

   - By applying different clustering algorithms on the same dataset and comparing their results using the metrics mentioned above, one can get a sense of which algorithm works best for that particular dataset.

6. **Elbow Method (specifically for K-means)**:
   
   - By plotting the number of clusters against the respective WCSS, an "elbow" point can be observed where the WCSS starts to level off. This point is often a good estimate for the optimal number of clusters. -->

When evaluating clustering models, it's crucial to remember that no single metric can tell the entire story. It's often beneficial to use a combination of internal and external metrics (if ground truth is available), visual inspection, and domain knowledge to assess the quality and usefulness of the clustering results.
