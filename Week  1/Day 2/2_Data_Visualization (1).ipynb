# Matplotlib

We'll now take an in-depth look at the Matplotlib package for visualization in Python. Matplotlib is a multi-platform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack. It was conceived by John Hunter in 2002, originally as a patch to IPython for enabling interactive MATLAB-style plotting via gnuplot from the IPython command line.


One of Matplotlib’s most important features is its ability to play well with many operating systems and graphics backends.
Matplotlib supports dozens of backends and output types, which means you can count on it to work regardless of which operating system you are using or which output format you wish.
This cross-platform, everything-to-everyone approach has been one of the great strengths of Matplotlib.

It has led to a large user base, which in turn has led to an active developer base and Matplotlib’s powerful tools and ubiquity within the scientific Python world.


[Matplotlib](http://matplotlib.org/) is the most common charting package, see its [documentation](http://matplotlib.org/api/pyplot_api.html) for details, and its [examples](http://matplotlib.org/gallery.html#statistics) for inspiration.

### Matplotlib Figure Anatomy

**Figure**
The figure is like a canvas where all your Axes (plots) are drawn. A figure can contain several Axes. For now, we will use only one.

**Axes**
This is what you think of as ‘a plot’.

The Axes contains two (or three in the case of 3D) Axis objects
* Each Axes has a title
* Each Axes can contain a legend

**Axis**
These are the number-line-like objects.

**Labels**
This is the "legend" of Axis. There are 2 labels for 2D plots the x_label and y_label.

**Ticks**
The ticks are the marks on the axis and ticklabels (strings labeling the ticks). There are two kind of ticks: major and minor ticks. By default they are automaticaly generated by the axis but they can be configured.

## Adjusting the Plot: Line Colors and Styles

The first adjustment you might wish to make to a plot is to control the line colors and styles.
The ``plt.plot()`` function takes additional arguments that can be used to specify these.
To adjust the color, you can use the ``color`` keyword, which accepts a string argument representing virtually any imaginable color.

````python
plt.plot(x, np.sin(x), color='blue')
plt.plot(x, np.cos(x), color='red')
plt.show()
````

The color can be specified in a variety of ways:

Note: If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

````python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.show()
````

Similarly, the line style can be adjusted using the ``linestyle`` keyword:

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
plt.show()
````

````python
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
plt.show()
````

If you would like to be extremely terse, these ``linestyle`` and ``color`` codes can be combined into a single non-keyword argument to the ``plt.plot()`` function:

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
plt.show()
````


There are many other keyword arguments that can be used to fine-tune the appearance of the plot; for more details: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

## Adjusting the Plot: Axes Limits

Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it's nice to have finer control.
The most basic way to adjust axis limits is to use the ``plt.xlim()`` and ``plt.ylim()`` methods:

````python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
plt.show()
````

The ``plt.axis()`` method allows you to set the ``x`` and ``y`` limits with a single call, by passing a list which specifies ``[xmin, xmax, ymin, ymax]``:

````python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
plt.show()
````

The ``plt.axis()`` method goes even beyond this, allowing you to do things like automatically tighten the bounds around the current plot:

````python
plt.plot(x, np.sin(x))
plt.axis('tight')
plt.show()
````

It allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in ``x`` is equal to one unit in ``y``:

````python
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.show()
````

For more information on axis limits and the other capabilities of the ``plt.axis`` method, refer to https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.axis.html

## Labeling Plots


As the last piece of this section, we'll briefly look at the labeling of plots: titles, axis labels, and simple legends.

Titles and axis labels are the simplest such labels—there are methods that can be used to quickly set them:

````python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
````

The position, size, and style of these labels can be adjusted using optional arguments to the function.
For more information, see the Matplotlib documentation : https://matplotlib.org/3.2.1/contents.html

**Legends**
When multiple lines are being shown within a single axes, it can be useful to create a plot legend that labels each line type. Again, Matplotlib has a built-in way of quickly creating such a legend. It is done via the (you guessed it) plt.legend() method. Though there are several valid ways of using this, I find it easiest to specify the label of each line using the label keyword of the plot function:


````python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.legend()
plt.show()
````

As you can see, the ``plt.legend()`` function keeps track of the line style and color, and matches these with the correct label.
More information on specifying and formatting plot legends can be found in https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

# Scatter Plots

Another commonly used plot type is the simple scatter plot, a close cousin of the line plot. Instead of points being joined by line segments, here the points are represented individually with a dot, circle, or other shape

## Scatter Plots with ``plt.plot``

In the previous section we looked at ``plt.plot``/``ax.plot`` to produce line plots.
It turns out that this same function can produce scatter plots as well:

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, '.', color='black');
plt.show()
````

The third argument in the function call is a character that represents the type of symbol used for the plotting. Just as you can specify options such as ``'-'``, ``'--'`` to control the line style, the marker style has its own set of short string codes. The full list of available symbols can be seen in the documentation of ``plt.plot``, or in Matplotlib's online documentation. 

Common ones shown below:

````python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
plt.show()
````

## Bar Plots

Bar plot can be plot using the `plot.bar()` function

````python
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]
````

**Customizing bar plots:** All customization that we discussed earlier in line plots can be applied to bar plots as well. Additionally, multiple bar plots can be combined in a single figure/plot

````python
# Look at index 4 and 6, which demonstrate overlapping cases.
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(x1, y1, label="Blue Bar", color='b')
plt.bar(x2, y2, label="Green Bar", color='g')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.legend()
plt.show()
````

Here’s a side-by-side bar chart. The bars are adjusted to avoid overlapping, making it easier to compare the two datasets.

````python
# Data
x1 = np.array([1, 3, 4, 5, 6, 7, 9])
y1 = np.array([4, 7, 2, 4, 7, 8, 3])

x2 = np.array([2, 4, 6, 8, 10])
y2 = np.array([5, 6, 2, 6, 2])

# Bar width to ensure side-by-side bars
bar_width = 0.4

# Adjust positions for side-by-side bars
x1_adjusted = x1 - bar_width / 2
x2_adjusted = x2 + bar_width / 2

# Plot
plt.bar(x1_adjusted, y1, width=bar_width, label="Blue Bar", color='b')
plt.bar(x2_adjusted, y2, width=bar_width, label="Green Bar", color='g')

plt.xlabel("Bar Number")
plt.ylabel("Bar Height")
plt.title("Side-by-Side Bar Chart Example")
plt.legend()
plt.show()
````

## Histograms

A simple histogram can be a great first step in understanding a dataset. Histograms can be plotted using the `plt.hist()` function from matplotlib

Let's first create some data

````python
data = np.random.randn(1000)
````

 `randn` generates an array of shape (d0, d1, ..., dn), filled with random floats sampled from a univariate “normal” (Gaussian) distribution of mean 0 and variance 1.

````python
data[:10]
````

````python
plt.hist(data);
plt.show()
````

The ``hist()`` function has many options to tune both the calculation and the display;
here's an example of a more customized histogram:

````python
plt.hist(data, bins=30, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none');
plt.show()
````

The ``plt.hist`` documentation (https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.hist.html) has more information on other customization options available.
Combination of ``histtype='stepfilled'`` along with some transparency ``alpha`` are  very useful when comparing histograms of several distributions:

````python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
````

````python
plt.hist(x1, histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x2, histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x3, histtype='stepfilled', alpha=0.3, bins=40);
plt.show()
````

## Visualization using Pandas dataframe

Pandas allows for visualization directly from a dataframe which provides easy to use interface, built on Matplotlib.



````python
# Let's create a random daataframe consisting of four columns from 1 to 100

import pandas as pd  #import pandas

# Sample DataFrame
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Sales": [150, 200, 250, 300, 350],
    "Profit": [20, 30, 50, 70, 90]
}

df = pd.DataFrame(data)
print(df)
````

## Line Plot

Using the dataframe generated above, Let's plot a line graph containing values from column A.

````python
df.plot(x="Year", y=["Sales", "Profit"], kind="line", marker='o', figsize=(8, 5), grid=True)
plt.title("Sales and Profit Over the Years")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend(title="Metrics")
plt.show()
````

## Bar Chart

````python
df.plot(x="Year", y="Sales", kind="bar", color="skyblue", figsize=(8, 5))
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
````

## Box Plot

````python
df[["Sales", "Profit"]].plot(kind="box", figsize=(8, 5))
plt.title("Box Plot for Sales and Profit")
plt.ylabel("Values")
plt.show()
````

## Pie Chart

````python
# Pie chart for sales
df["Sales"].plot(kind="pie", labels=df["Year"], autopct="%1.1f%%", figsize=(8, 8), startangle=90, colormap="Pastel1")
plt.title("Proportion of Sales by Year")
plt.ylabel("")  # Hide the y-axis label
plt.show()
````

Seaborn also tries to promote techniques that are powerful but less familiar, such as calculating and plotting the empirical cumulative distribution function of the data:

### Specialized plots for categorical data
Several specialized plot types in seaborn are oriented towards visualizing categorical data. They can be accessed through ``catplot()``. These plots offer different levels of granularity. At the finest level, you may wish to see every observation by drawing a “violin” plot: a scatter plot that adjusts the positions of the points along the categorical axis so that they don’t overlap:

````python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
plt.show()
````

Visualizing the multidimensional relationships among the samples is as easy as calling ```sns.pairplot:```

````python
# using the iris dataset
df = sns.load_dataset("iris")

df
````

````python
sns.pairplot(df, hue='species', height=2.5);
plt.show()
````

## Faceted histograms
Sometimes the best way to view data is via histograms of subsets. Seaborn's ```FacetGrid``` makes this extremely simple. We'll take a look at some data that shows the amount that restaurant staff receive in tips based on various indicator data:

````python
df_tips = sns.load_dataset('tips')
df_tips.head()
````

````python
df_tips['tip_pct'] = 100 * df_tips['tip'] / tips['total_bill']

df_tips
````

````python
grid = sns.FacetGrid(df_tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
plt.show()
````

## Heatmap


A heat map (or heatmap) is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving obvious visual cues to the reader about how the phenomenon is clustered or varies over space.

````python
flights=sns.load_dataset("flights")

flights
````

````python
# pd.pivot(index=None, columns=None, values=None)
flights = flights.pivot(index="month", columns="year", values="passengers")
flights
````

````python
sns.heatmap(flights,linewidths=.5,cmap="YlGnBu")
plt.show()
````
