### 1. Descriptive statistics and measures of central tendency.

**Exercise:** Given the list of numbers below, calculate the mean, median, and
mode.
`numbers = [10, 20, 30, 20, 20, 30, 40, 50, 60, 70, 80, 90, 100]`

```python
from scipy import stats
import numpy as np

numbers = [10, 20, 30, 20, 20, 30, 40, 50, 60, 70, 80, 90, 100]

mean = np.mean(numbers)
median = np.median(numbers)
mode = stats.mode(numbers)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

### 2. Measures of variability and distribution.

**Exercise:** Using the same list of numbers from the previous exercise,
calculate the variance, and standard deviation.

```python
import numpy as np

numbers = np.array([10, 20, 30, 20, 20, 30, 40, 50, 60, 70, 80, 90, 100])

variance = np.var(numbers)
std_dev = np.std(numbers)

print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

### 3. Hypothesis testing and significance testing.

**Exercise:** Letâ€™s say we have measured the heights (in cm) of two different
groups of plants grown under different light conditions. Perform a t-test to
see if the difference between these groups is statistically significant.
`group1 = [180, 190, 175, 185, 195, 200, 210, 205]` `group2 = [170, 165, 180, 175,
185, 190, 180, 175]`

```python
from scipy import stats

group1 = [180, 190, 175, 185, 195, 200, 210, 205]
group2 = [170, 165, 180, 175, 185, 190, 180, 175]

t_stat, p_val = stats.ttest_ind(group1, group2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

### 4. Confidence intervals.

**Exercise:** Using the data from group1 above, calculate the 95% confidence
interval for the mean plant height.

```python
import scipy.stats as stats

group1 = np.array([180, 190, 175, 185, 195, 200, 210, 205])

confidence = 0.95
mean = np.mean(group1)
sem = stats.sem(group1)  # standard error of the mean
ci = stats.t.interval(confidence, len(group1)-1, loc=mean, scale=sem)

print(f"95% Confidence Interval: {ci}")
```

The 95% confidence interval means that if we were to repeat our sampling many times, 95% of the time our sample mean would fall within this interval.
